**1.""Iris Flowers Dataset""**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Load the dataset
df = pd.read_csv("/content/iris.csv")
df
```

```python
# 1. Data Analysis & Summarization
print("Summary Statistics:")
print(df.describe())
```

```python
# 2. Statistical Analysis
sns.histplot(x = 'setosa',data = df)
plt.title('Distribution of setosa')
plt.show()
```

```python
sns.boxplot(df['setosa'])
plt.title('Boxplot of setosa by Species')
plt.show()
```

```python
# 3. Correlation Coefficients
corr = df[['setosa','versicolor','virginica']].corr()
print("Correlation Matrix:")
print(corr)
```

```python
# 4. NULL Values
print("NULL Values in Each Column:")
print(df.isnull().sum())
```

```python
# Re-check NULL Values
print("NULL Values After Imputation:")
print(df.isnull().sum())
```

```python
# Handle NULL values (if any) in iris flower dataset there is no null value and objet is str so the fillna function might give error as it cannot convert str to numeric values
# Example of filling NULL values with mean
df['setosa'].fillna(df['setosa'].mean(), inplace=True)
df
```

**3."""Ionosphere dataset"""**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Abalone dataset
df = pd.read_csv('/content/Ionosphere.csv')
df
```

```python
# Data Analysis & Summarization
print("Summary Statistics for Ionosphere Dataset:")
print(df.describe())
```

```python
# Statistical Analysis
sns.histplot(df["V30"]) # Replace 'V30' with actual column names
plt.title('Distribution of V30')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation Coefficients
correlation_matrix = df.select_dtypes(include='number').corr()
print("Correlation Matrix for Abalone Dataset:")
print(correlation_matrix)
```

```python
# NULL Values
print("NULL Values in Each Column:")
print(df.isnull().sum())
df['V25'].fillna(df['V26'].mean(), inplace=True) # Replace 'V25', 'V26' with actual column names
print("NULL Values After Imputation:")
print(df.isnull().sum())
```

**4."""Wheat-seeds dataset"""**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Abalone dataset
df = pd.read_csv('/content/wheat-seeds (1).csv')
df
```

```python
# Data Analysis & Summarization
print("Summary Statistics for Wheat Seeds Dataset:")
print(df.describe())
```

```python
# Statistical Analysis
sns.histplot(df["perimeter"])
plt.title('Distribution of perimeter')
plt.xlabel('perimeter')
plt.ylabel('frequency')
plt.show()
```

```python
sns.boxplot(x='perimeter', y='area', data=df)
plt.title('Boxplot of area by perimeter')
plt.xlabel('perimeter')
plt.ylabel('area')
plt.show()
```

```python
# Correlation Coefficients
correlation_matrix = df.select_dtypes(include='number').corr()
print("Correlation Matrix for Wheat Seeds Dataset:")
print(correlation_matrix)
```

```python
# NULL Values
print("NULL Values in Each Column:")
print(df.isnull().sum())
df['area'].fillna(df['perimeter'].mean(), inplace=True)
print("NULL Values After Imputation:")
print(df.isnull().sum())
```
